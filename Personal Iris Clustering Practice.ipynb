```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set()
```

```python
raw_data=pd.read_csv("iris_dataset.csv")
raw_data.head(120)
```

### Clustering without standardization

```python
x=raw_data.copy()
x
```

```python
kmeans=KMeans(3)
kmeans.fit(x)
```

```python
data_with_clusters=x.copy()
data_with_clusters['Clusters']=kmeans.fit_predict(x)
data_with_clusters
```

```python
plt.scatter(data_with_clusters['sepal_length'],data_with_clusters['sepal_width'],
        c=data_with_clusters['Clusters'],cmap='rainbow')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Sepal clustering without standardizing")
plt.show()
```

```python
plt.scatter(data_with_clusters['petal_length'],data_with_clusters['petal_width'],
           c=data_with_clusters['Clusters'],cmap='rainbow')
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Petal clustering without standardizing")
plt.show()
```

### Clustering after standardization

```python
from sklearn import preprocessing

x_scaled=preprocessing.scale(x)
x_scaled
```

```python
kmeans_scaled=KMeans(3)
kmeans_scaled.fit(x_scaled)
```

```python
data_with_clusters_scaled=x.copy()
data_with_clusters_scaled['Clusters']=kmeans_scaled.fit_predict(x_scaled)
data_with_clusters_scaled
```

```python
plt.scatter(data_with_clusters_scaled['sepal_length'],data_with_clusters_scaled['sepal_width'],
           c=data_with_clusters_scaled['Clusters'],cmap='rainbow')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Standardized sepal clusters")
plt.show()
```

```python
plt.scatter(data_with_clusters_scaled['petal_length'],data_with_clusters_scaled['petal_width'],
           c=data_with_clusters_scaled['Clusters'],cmap='rainbow')
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Standardized petal clusters")
plt.show()
```
